```python
# Function to add all odd numbers from 1 to n
def fun1(n):
    sum = 0

    for i in range(1, n + 1, 2):
        sum += i

    return sum
    # Function ends here

n = int(input("Enter a number: "))
if n < 1:
    print("Please enter a larger value")
else:
    res = fun1(n)
    print("The sum of odd numbers from 1 to ", n , "is: ", res)
```

```python
# Helper function to check if num is prime
def helper(num):
    if num < 1:
        return False
    for i in range(2, int(num/2) + 1):

        if num % i == 0:
            return False

    return True
    # Helper function ends

# Function to find sum of prime numbers from 1 to n
def fun2(n):
    sum = 0

    for i in range(1, n + 1):

        if helper(i):
            sum += i

    return sum
    # Fun2 ends here

num = int(input("Enter a number: "))
if num < 2:
    print("Enter a larger value")
else:
    result = fun2(num)
    print("The sum of prime numbers from 1 to ", num , "is: ", result)
```
