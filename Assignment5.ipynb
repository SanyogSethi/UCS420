Assignment 5.2

```python
num = int(input("Enter a number: "))
sum = 0
for i in range(1, num+1):
  if (i % 7 == 0 and i % 9 == 0):
    sum += i
print("The sum of numbers divisible by 7 & 9 from 1 to", num ," is: ", sum)
```

Assignment 5.3

```python
numm = int(input("Enter a number: "))
summ = 0

for i in range(1, numm+1):
  for j in range(2, int(i/2) +1):
    if( i % j == 0):
      break
  else:
    summ += i
print("The sum of prime numbers from 1 to " , numm , " is: ",summ)
```
